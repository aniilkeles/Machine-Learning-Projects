```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# median , quartile(25%,50%,75%) , min , max 1.5 *IQR ,outlier
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])
plt.figure(figsize=(6,5)) # figsize la daha büyük ya da daha küçük yapabiliriz box u
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data Value")
plt.grid(True)
plt.show()
# 25 ve 75 lik yerler gösterilir ve 1.5 katına kadar olan yerlerde whisker(kedi bıyığı) olarak gösterilir 100 de burda outlier(çıkıntı)
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
plt.figure(figsize=(13,8))

plt.subplot(1,2,1)
sns.boxplot(x="class",y="age",data=df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")
plt.subplot(1,2,2)
sns.boxplot(x="class",y="fare",data=df)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("fare")
plt.tight_layout()
plt.show()
```
