```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/Dataset/athlete_events.csv')
```

```python
data.describe()
```

```python
data.head()
```

```python
data.info()
```

```python
plt.scatter("Height","Weight",data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlete Height vs Weight")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height",y="Weight",data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Atheletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height",y="Weight",hue="Sex",data=data) # hue bize hangi kolonla ilgili bir grselletirme yapma imkan sunar ayrt etme imkan sunar kategorik olmal setiimiz kolon ama ok etiketlide olmamal
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Atheletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height",y="Weight",hue="Sex",style="Medal",data=data) # hue bize hangi kolonla ilgili bir grselletirme yapma imkan sunar ayrt etme imkan sunar kategorik olmal setiimiz kolon ama ok etiketlide olmamal
plt.xlabel("Height of Athletes")                                         # style da bize iliglindiimiz kolona ait featurelar daha belirgin klarak yorumlamamza yardmc olur
plt.ylabel("Weight of Athletes")
plt.title("Atheletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x="Height",y="Weight",hue="Sex",style="Medal",size="Age",data=data) # hue bize hangi kolonla ilgili bir grselletirme yapma imkan sunar ayrt etme imkan sunar kategorik olmal setiimiz kolon ama ok etiketlide olmamal
plt.xlabel("Height of Athletes")                                         # style da bize iliglindiimiz kolona ait featurelar daha belirgin klarak yorumlamamza yardmc olur
plt.ylabel("Weight of Athletes")                                         # size age ' e gre mesela bir sralama yapt pointlere
plt.title("Atheletes Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height",y="Weight",hue="Sex",data=data)
plt.xlabel("Heights of Athletes")
plt.ylabel("Weights of Athletes")
plt.title("Athletes Weights and Heights")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height",hue="Sex",data=data)
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
```

```python
sns.set_style("white")
sns.displot(x="Height",hue="Sex",data=data,kind ="kde")
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
```

```python
sns.barplot(x="Medal",y="Height",hue="Sex",data=data)
plt.title("Medals by Height")
plt.show()
```

```python
sns.catplot(x="Medal",y="Height",hue="Sex",col="Season",data=data)
plt.show()
```
