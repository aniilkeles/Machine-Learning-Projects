```python
X = df.drop("Outcome",axis=1)
y = df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=15)
```

```python
cols_row_drop = ["Glucose","BloodPressure","BMI"]

for col in cols_row_drop:
  X_train = X_train[X_train[col] != 0]
  X_test = X_test[X_test[col] != 0]
```

```python
y_train = y_train.loc[X_train.index]
y_test = y_test.loc[X_test.index]
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
ada = AdaBoostClassifier()
ada.fit(X_train,y_train)
y_pred = ada.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
# adaboostclassifier vs other types of classification models
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
def calculate_metrics(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    return acc, precision, recall, f1
```

```python
models = {
          "Logistic Regression" : LogisticRegression(),
          "Naive Bayes : " : GaussianNB(),
          "KNN Classifier : " : KNeighborsClassifier(),
          "Support Vector Classifier : " : SVC(),
          "DecisionTree Classifier : " : DecisionTreeClassifier(),
          "RandomForest Classifier : " : RandomForestClassifier()
}
```

```python
for name,model in models.items():
    model.fit(X_train,y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)


    # TRAIN METRICS
    train_acc, train_prec, train_rec, train_f1 = calculate_metrics(
        y_train, y_train_pred
    )

    # TEST METRICS
    test_acc, test_prec, test_rec, test_f1 = calculate_metrics(
        y_test, y_test_pred
    )

    print(name)

    print("Evaluation for Training Set")
    print("Accuracy : ", train_acc)
    print("Precision : ", train_prec)
    print("Recall : ", train_rec)
    print("F1 Score : ", train_f1)

    print("------------------------------------")

    print("Evaluation for Test Set")
    print("Accuracy : ", test_acc)
    print("Precision : ", test_prec)
    print("Recall : ", test_rec)
    print("F1 Score : ", test_f1)

    print("--------------------------------------")

    print("Classification Report (Test)")
    print(classification_report(y_test, y_test_pred))

    print("Confusion Matrix (Test)")
    print(confusion_matrix(y_test, y_test_pred))

    print("\n")
```

```python
# hyperparameter-tuning for other types of classification models
```

```python
knn_params ={"n_neighbors" : [2,3,10,20,40,50]}
rf_params = {
          "max_depth" : [5,8,10,15,None],
          "max_features" : ["sqrt","log2",5,7,10],
          "min_samples_split" : [2,8,12,15,20],
          "n_estimators" : [100,200,500,1000]
}
svc_params = {
    "C": [0.1, 1, 10, 100],
    "gamma": ["scale", 0.1, 0.01, 0.001],
    "kernel": ["rbf", "linear"]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv_models = [
    ("KNN",KNeighborsClassifier(),knn_params),
    ("RF",RandomForestClassifier(random_state=42),rf_params),
    ("SVC",SVC(),svc_params)
]
```

```python
for name,model,params in randomcv_models:
  randomcv = RandomizedSearchCV(estimator=model,param_distributions=params,n_iter=30,cv=5,n_jobs=-1,scoring="accuracy",random_state=42)
  randomcv.fit(X_train,y_train)
  print("best params for : ",name, randomcv.best_params_)
```
