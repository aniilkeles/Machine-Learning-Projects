```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
df[["sex","embarked","embark_town"]].isna().sum()
```

```python
df = df.dropna(subset=['embark_town']) # df = df["embark_town"].dropna() df.dropna(subset=['embark_town']) similar with first sentence
```

```python
df
```

```python
df[["sex","embarked","embark_town"]].isna().sum()
```

```python
# one hot encoding
```

```python
df['sex'].value_counts()
```

```python
df_onehot = pd.get_dummies(df,columns = ["sex","embark_town"],drop_first=True) # drop_first = true ilk değeri bizim için siler
```

```python
df_onehot
```

```python
# label encoder
```

```python
label_encoder = LabelEncoder()
```

```python
df_label = df.copy()
```

```python
df_label
```

```python
df_label["sex"]=label_encoder.fit_transform(df_label["sex"])
```

```python
df_label
```

```python
df
```

```python
# ordinal encoder
```

```python
df_ordinal = df.copy()
```

```python
class_order = ["Third","Second","First"]
```

```python
ordinal_encoder = OrdinalEncoder(categories=[class_order])
```

```python
df_ordinal["class"]=ordinal_encoder.fit_transform(df_ordinal[["class"]])
```

```python
df_ordinal
```

```python
df
```

```python
fig,axes = plt.subplots(1,3,figsize=(15,5))
df["sex"].value_counts().plot(kind="bar",ax=axes[0],title="Original Categorical")
df_label["sex"].value_counts().plot(kind="bar",ax=axes[1],title="Label Encoded")
df_onehot["sex_male"].value_counts().plot(kind="bar",ax=axes[2],title="One Hot Encoded")
```
