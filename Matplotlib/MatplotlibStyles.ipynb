```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
data1 = np.linspace(0,10,20)
```

```python
data2 = data1**2
```

```python
data1
```

```python
plt.subplots() # subplots bize hazır bir figür ve axis döndürür tuple içinde
```

```python
my_fig,my_axes = plt.subplots()
```

```python
type(my_fig)
```

```python
type(my_axes)
```

```python
my_fig,my_axes = plt.subplots()
my_axes.plot(data1,data2,color = "#1E0703",alpha = 0.1) #alpha shows transperency to us
my_axes.plot(data2,data1)
plt.show()
```

```python
new_fig,new_axes = plt.subplots()

new_axes.plot(data1,data1 + 2,color ="red",linewidth=1.0)
new_axes.plot(data1,data1 + 3,color ="yellow",linewidth=4.0)
new_axes.plot(data1,data1 + 4,color ="blue",linestyle="-.")
new_axes.plot(data1,data1 + 5,color ="green",linestyle="--",marker="o",markersize=5,markerfacecolor="black")
plt.show()
```
