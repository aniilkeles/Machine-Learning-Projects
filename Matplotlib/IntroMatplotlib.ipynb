```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list= [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list,weight_list,"g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
#numpy
```

```python
np_age_list = np.array(age_list)
```

```python
np_weight_list = np.array(weight_list)
```

```python
np_age_list
```

```python
plt.plot(np_age_list,np_weight_list,"r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
# linspace => it produces numbers which are produced by us but eqaully ratio
```

```python
num_arr1 = np.linspace(1,10,20)
```

```python
num_arr1
```

```python
num_arr2= num_arr1**3
```

```python
num_arr2
```

```python
plt.plot(num_arr1,num_arr2)
plt.show()
#EXPONENTIAL (STEL)
```

```python
plt.plot(num_arr1,num_arr2,"r*-")
plt.show()
```

```python
plt.subplot(1,2,1) # => (1.row,1.column,1.graphic)
plt.plot(num_arr1,num_arr2,"r*-")
plt.subplot(1,2,2)
plt.plot(num_arr2,num_arr1,"g--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.3,0.3]) # => 0.1 0.1 ler soldan ve alttan boluu ifade edip , 0.3ler de ykseklik ve genilii ifade eder
figure_axes.plot(num_arr1,num_arr2,"g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=100) # dpi znrlk

new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(num_arr1, num_arr1**2,label = "numpy array **2")
new_axes.plot(num_arr1, num_arr1**3,label = "numpy array **3")
new_axes.legend()
plt.show()
```
