```python
X = df.drop('class',axis=1)
y=df["class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=15)
```

```python
X_train.shape
```

```python
# ordinal encoding
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
df.head()
```

```python
categorical_cols = ["buying","maintenance","lug_boot","safety"]
numerical_cols = ["doors","persons"]
ordinal_encoder = OrdinalEncoder(categories=[
    ["low","med","high","vhigh"], # buying
    ["low","med","high","vhigh"], # maintenance
    ["small","med","big"], # lug_boot
    ["low","med","high"] # safety
])


# when we want to obtain difference aprroaches to OrdinalEncoder,we can use pipeline.

preprocessor = ColumnTransformer(transformers=[
   ('transformation_name',ordinal_encoder,categorical_cols)
    # ('asdsa',one_hot_encoder,one_hot_cols)
    # ('dsasd',label_encoder,label_cols)
] , remainder = 'passthrough'

                                 )

X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
# if we don't assign to remainder passthrough , the columns which are we didn't assign as encoder , it can drop
# This method provides encoding to columns
```

```python
pd.DataFrame(X_train_transformed)[5].unique()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=0)
```

```python
tree_model.fit(X_train_transformed,y_train)
```

```python
y_pred = tree_model.predict(X_test_transformed)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
plt.figure(figsize=(12,8))

from sklearn import tree

column_names = categorical_cols+numerical_cols

tree.plot_tree(tree_model.fit(X_train_transformed,y_train),feature_names=column_names)
```

```python
 # hyperparameter tuning
```

```python
param = {
    "criterion" : ["gini","entropy","log_loss"],
    "splitter" : ["best","random"],
    "max_depth" : [1,2,3,4,5,16,None],
    "max_features" : ["sqrt","log2",None]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=DecisionTreeClassifier(), param_grid = param, cv = 5, scoring="accuracy")
```

```python
import warnings
warnings.filterwarnings("ignore")
grid.fit(X_train_transformed,y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test_transformed)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
tree_model_new = DecisionTreeClassifier(max_depth=16,criterion='entropy',max_features=None,splitter='best')
```

```python
tree_model_new.fit(X_train_transformed,y_train)
y_pred=tree_model_new.predict(X_test_transformed)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
